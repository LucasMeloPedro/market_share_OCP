```python
#importando biblioteca de manipulação de dados: Pandas
import pandas as pd
```

```python
# Lendo e trazendo para o notebook o conjunto de dados
df = pd.read_csv('certificados_produtos.csv', encoding = 'utf-16', sep = ';', on_bad_lines='skip')
df.head(10)
```

```python
# Verificando tamanho da tabela
df.shape
```

```python
# Verificando valores ausentes
df.isna().sum()
```

```python
# Verificando valores ausents na coluna número do certificado
df[df['NUM_CERTIFICADO'].isna()]
```

```python
# Gerando novo dataframe removendo algumas colunas do conjunto inicial (alto número de dados ausentes)
dados = df.drop(columns = ['EMPRESA_NOME_FANTASIA', 'EMPRESA_BAIRRO', 'EMPRESA_TELEFONE', 'EMPRESA_EMAIL'])
dados.head()
```

```python
dados.isna().sum()
```

```python
dados.shape
```

```python
dados = dados.dropna(subset = ['STATUS', 'NUM_CERTIFICADO', 'CERTIFICADOR_SIGLA'])
dados.isna().sum()
```

```python
dados[dados['EMPRESA_RAZAO_SOCIAL'].isna()]
```

```python
# Substituindo os valores ausentes pela frase "Não informado" para as demais colunas que não foram tratadas nos passos anteriores
dados.fillna(value = 'Não Informado', inplace = True)
dados.isna().sum()
```

```python
dados.head()
```

```python
# Remoção de linhas duplicadas (certificados duplicados)
dados = dados.drop_duplicates(subset = 'NUM_CERTIFICADO')
dados.shape
```

```python
dados.dtypes
```

```python
# Transformando datas com vencimento acima do aceitável pelo datetime
date_out = dados['DATA_VALIDADE'].str[6:10].astype(int) >= 2262
dados.loc[date_out, 'DATA_VALIDADE'] = '30/12/2199'
```

```python
dados[dados['DATA_VALIDADE'] == '12/03/3019']
```

```python
# Transformação de colunas de tipo objeto para tipo data 
dados['DATA_EMISSAO'] = pd.to_datetime(dados['DATA_EMISSAO'], format='%d/%m/%Y')
dados['DATA_VALIDADE'] = pd.to_datetime(dados['DATA_VALIDADE'], format='%d/%m/%Y')
```
